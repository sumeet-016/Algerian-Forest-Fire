## About Dataset
The dataset includes 244 instances that regroup a data of two regions of Algeria, namely the Bejaia region located in the northeast of Algeria and the Sidi Bel-abbes region located in the northwest of Algeria.

122 instances for each region.

The period from June 2012 to September 2012. The dataset includes 11 attributes and 1 output attribute (class) The 244 instances have been classified into fire (138 classes) and not fire (106 classes) classes.

## Attribute Informations 

1. Date : (DD/MM/YYYY) Day, month ('june' to 'september'), year (2012) Weather data observations
2. Temp : temperature noon (temperature max) in Celsius degrees: 22 to 42
3. RH : Relative Humidity in %: 21 to 90
4. Ws :Wind speed in km/h: 6 to 29
5. Rain: total day in mm: 0 to 16.8 FWI Components
6. Fine Fuel Moisture Code (FFMC) index from the FWI system: 28.6 to 92.5
7. Duff Moisture Code (DMC) index from the FWI system: 1.1 to 65.9
8. Drought Code (DC) index from the FWI system: 7 to 220.4
9. Initial Spread Index (ISI) index from the FWI system: 0 to 18.5
10. Buildup Index (BUI) index from the FWI system: 1.1 to 68
11. Fire Weather Index (FWI) Index: 0 to 31.1
12. Classes: two classes, namely Fire and not Fire

## Data Cleaning

```python
data[data.isnull().any(axis=1)]
```

```python
data.dropna(subset=['day'], inplace=True)
```

The dataset is converted into two sets based on Region from 123th index, we can make a new column based on Region  

1. "Bejaia Region Dataset"
2. "siddi-Bel Abbes Region Dataset"

Add new column with region

```python
data.loc[:123, "Region"]=0
data.loc[122:, "Region"]=1
data.info()
```

```python
data.head()
```

```python
data[['Region']] = data[['Region']].astype(int)
data.head()
```

```python
data.isnull().sum()
```

```python
data = data.dropna().reset_index(drop=True)
data.isnull().sum()
```

```python
data.iloc[[122]]
```

```python
data = data.drop(122).reset_index(drop=True)
```

```python
data.iloc[[122]]
```

```python
data.columns
```

```python
data.columns = data.columns.str.strip()
data.columns
```

Changes the required columns as int data type

```python
data[['month', 'year', 'day', 'Temperature', 'RH', 'Ws']] = data[['month', 'year', 'day', 'Temperature', 'RH', 'Ws']].astype(int)
data.info()
```

Changing rest of the columns into float datatype

```python
objects = [features for features in data.columns if data[features].dtypes == 'O']
objects
```

```python
for i in objects:
    if i != 'Classes':
        data[i] = data[i].astype(float)

data.info()
```

```python
data.to_csv('Algerian_forest_fires_update_dataset.csv', index=False)
```
